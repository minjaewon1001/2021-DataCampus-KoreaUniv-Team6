```python
# https://yobbicorgi.tistory.com/29
# https://hnm1.tistory.com/entry/%ED%8C%8C%EC%9D%B4%EC%8D%AC-%EA%B5%AC%EA%B8%80-%EC%9D%B4%EB%AF%B8%EC%A7%80-%ED%81%AC%EB%A1%A4%EB%A7%81-%EA%B0%95%EC%9D%98-%EC%9D%B4%EB%AF%B8%EC%A7%80-%ED%81%AC%EB%A1%A4%EB%A7%81-%EB%B0%A9%EB%B2%95-Selenium-%EC%82%AC%EC%9A%A9%EB%B2%95

from selenium import webdriver 
from selenium.webdriver.common.keys import Keys
import time
import os
import urllib.request

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
        
keyword=input("검색할 키워드를 입력하세요 : ")
createFolder('F:/데이터캠퍼스/'+keyword+'_img_download')

b = int(input("개수 : "))

driver = webdriver.Chrome('F:/데이터캠퍼스/chromedriver.exe')
driver.implicitly_wait(3)


# =============================================================================
# 구글 이미지 검색 접속 및 검색어 입력
# =============================================================================
print(keyword, '검색')
driver.get('https://www.google.co.kr/imghp?hl=ko')

Keyword=driver.find_element_by_xpath('//*[@id="sbtc"]/div/div[2]/input')
Keyword.send_keys(keyword)

driver.find_element_by_xpath('//*[@id="sbtc"]/button').click()


# =============================================================================
# 스크롤
# =============================================================================
print(keyword+' 스크롤 중 .............')
elem =  driver.find_element_by_tag_name("body")
for i in range(60):
    elem.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.1)
    
try:
    driver.find_element_by_xpath('//*[@id="islmp"]/div/div/div/div[1]/div[4]/div[2]/input').click()
    for i in range(60):
        elem.send_keys(Keys.PAGE_DOWN)
        time.sleep(0.1)
except:
    pass

# 다시 페이지 올리기
for i in range(60):
    elem.send_keys(Keys.PAGE_UP)
    time.sleep(0.1)


# =============================================================================
# 이미지 개수
# =============================================================================
links=[]
images = driver.find_elements_by_css_selector("img.rg_i.Q4LuWd")
for image in images:
    if image.get_attribute('src')!=None:
        links.append(image.get_attribute('src'))

print(keyword+' 찾은 이미지 개수:',len(links))
time.sleep(2)

```

```python

# =============================================================================
# 이미지 다운로드
# =============================================================================
img_links=[]
count = 0

for image in images:
    try:
        image.click()
        time.sleep(2)
        
        imgUrl = driver.find_elements_by_css_selector("img.n3VNCb")
        
        if imgUrl[0].get_attribute('src')!=None:
            img_links.append(imgUrl[0].get_attribute('src'))
        print(len(img_links)+'번째 링크 찾는 중........')
        
        count += 1
        if count == b:
            break
            
    except:
        pass

    
for k,i in enumerate(img_links):
    url = i
    start = time.time()
    urllib.request.urlretrieve(url, "F:/데이터캠퍼스/"+keyword+"_img_download/"+keyword+"_"+str(k)+".jpg")
    print(str(k+1)+'/'+str(len(img_links))+' '+' 다운로드 중....... Download time : '+str(time.time() - start)[:5]+' 초')
    
driver.close()
```
